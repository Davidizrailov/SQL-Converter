```python
dot = Digraph()
dot.graph_attr.update({'rankdir': 'LR'})
```

```python
f = open('jta_packages.json', 'r')
import json
data = json.load(f)
print(data)
graph = {}
ts = []
for i, row in enumerate(data):
  graph[row['name']] = row['tables_accessed']
  ts.extend(row['tables_accessed'])
print(graph)
```

```python
import re
f = open('jta_packages.html', 'r')
text = f.read()
pattern = r"(?is)<table.*?>.*?</table>"
tables = re.findall(pattern, text, re.DOTALL)
  
```

```python
for i, table in enumerate(tables):
  dot.node(str(data[i]['name']), label="<"+table.strip()+">", shape='plaintext')
```

```python
dot.render(filename='a', format='svg')
```

```python
ts = list(set(ts))
prev = None
for i in ts:
  dot.node(i)
  if prev:
    dot.edge(prev, i, style='invis', weight='50')
  prev = i
prev = None
for key, value in graph.items():
  # if prev:
  #   dot.edge(prev, key, style='invis')
  # prev = key
  for x in value:
    dot.edge(key, x)
```
