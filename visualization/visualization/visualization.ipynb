```python
from graphviz import *
```

```python
dot = Graph('JTA codebase diagram')
dot
```

```python
dot = Graph()
```

```python
import re
f = open('jta_packages.html', 'r')
text = f.read()
pattern = r"(?is)<table.*?>.*?</table>"
tables = re.findall(pattern, text, re.DOTALL)
  
```

```python
for table in tables:
  dot.node('', label="<"+table+">", shape='plaintext')
```

```python
dot.render(filename='a', format='png')
```
